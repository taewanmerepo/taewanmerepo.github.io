# Matplotlib 한글 설정

## 한글 폰

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

t = np.arange(0.0, 2.0, 0.01)
s = 1 + np.sin(2 * np.pi * t)

plt.plot(t, s)
plt.title("한글테스트")
plt.show()
```

```python
import matplotlib
import matplotlib.pyplot as plt
from matplotlib import rc, font_manager
import numpy as np

rc('font', family="NanumGothic")
t = np.arange(0.0, 2.0, 0.01)
s = 1 + np.sin(2 * np.pi * t)

plt.plot(t, s)
plt.title("한글테스트")
plt.show()
```
